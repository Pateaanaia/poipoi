```python
import pandas as pd
import os
import datetime
import re
```

```python
end_date = datetime.date(year=2024, 31-08-2024)

date = datetime.date(year=2024, month=10, day=22)
date.strftime("%m-%d-%Y")
```

```python
csv_list = [f for f in os.listdir('./historique_lit') if os.path.isfile(os.path.join('./historique_lit', f))]
```

```python
def csv_name2date(texte):
    pattern = r'(\d{2})-(\d{2})-(\d{4})'
    match = re.search(pattern, texte)
    date_str = match.group(0)
    date_obj = datetime.datetime.strptime(date_str, '%d-%m-%Y')
    return date_obj
```

```python
start_date = csv_name2date(csv_list[0])
end_date = csv_name2date(csv_list[-1])
```

```python
current_date = start_date
while current_date <= end_date:
    # Affiche la date au format jour-mois-année
    print(current_date.strftime("%d-%m-%Y"))
    # Ajoute un jour
    current_date += datetime.timedelta(days=1)
```
