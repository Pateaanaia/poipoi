```python
def csv_name2date(texte):
    pattern = r'(\d{2})-(\d{2})-(\d{4})'
    match = re.search(pattern, texte)
    date_str = match.group(0)
    date_obj = datetime.datetime.strptime(date_str, '%d-%m-%Y')
    return date_obj
```

```python
def reduce_day(liste_strings, date):
    print("./historique_lit/LITS_OCCUPES_" + date)
    daily_files = [s for s in liste_strings if "LITS_OCCUPES_" + date in s]
    df1 = pd.DataFrame({ "PATIENT": []})
    for file_path in daily_files:
        total_file_path = "./historique_lit/" + file_path
        if os.path.getsize(total_file_path) != 0:
            try:
                dfi = pd.read_csv(total_file_path, encoding='cp1252', sep=";", on_bad_lines='skip')
            except UnicodeDecodeError:
                with open(total_file_path, "rb") as f:
                    result = chardet.detect(f.read())
                dfi = pd.read_csv(total_file_path, encoding=result['encoding'], sep=";", on_bad_lines='skip')            
            df1 = pd.concat([df1, dfi], axis=0)
    df1["DATE"] = datetime.datetime.strptime(date, '%d-%m-%Y')
    df1 = df1[["DATE", "PATIENT"]].drop_duplicates()
    return df1
#df = reduce_day(csv_list, "24-02-2024")
#df
```

```python
def extract_unique_dates(file_names: List[str]) -> List[str]:
    dates = set()  # Utilisation d'un ensemble pour éviter les doublons

    for file_name in file_names:
        # On extrait la partie date en supposant que le format est toujours le même
        parts = "-".join(file_name.split('_')[2].split('-')[0:3])  # Extraction de la date à partir du nom du fichier
        dates.add(parts)  # Ajoute la date sans les doublons grâce à l'ensemble
    
    return list(dates)  # Conversion en liste

# Exemple d'utilisation
unique_dates = extract_unique_dates(csv_list)
print(unique_dates)
```
