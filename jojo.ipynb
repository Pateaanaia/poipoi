```python
import pandas as pd
import os
import datetime
import re
```

```python
csv_list = [f for f in os.listdir('./historique_lit') if os.path.isfile(os.path.join('./historique_lit', f))]
```

```python
def csv_name2date(texte):
    pattern = r'(\d{2})-(\d{2})-(\d{4})'
    match = re.search(pattern, texte)
    date_str = match.group(0)
    date_obj = datetime.datetime.strptime(date_str, '%d-%m-%Y')
    return date_obj
```

```python
start_date = csv_name2date(csv_list[0])
end_date = csv_name2date(csv_list[-1])
print(start_date)
print(end_date)
```

```python
def reduce_day(liste_strings, date):
    print("./historique_lit/LITS_OCCUPES_" + date)
    daily_files = [s for s in liste_strings if "LITS_OCCUPES_" + date in s]
    df1 = pd.DataFrame({ "PATIENT": []})
    for file_path in daily_files:
        dfi = pd.read_csv("./historique_lit/" + file_path, encoding='cp1252', sep=";", on_bad_lines='skip')
        df1 = pd.concat([df1, dfi], axis=0)
    df1["DATE"] = datetime.datetime.strptime(date, '%d-%m-%Y')
    df1 = df1[["DATE", "PATIENT"]].drop_duplicates()
    return df1
#df = reduce_day(csv_list, "24-02-2024")
#df
```
